```python
from model_runner import ModelRunner
from image_helper import ImageHelper
from GaussianNoise import AddGaussianNoise
import torch
from torchvision import transforms
```

```python
model_runner = ModelRunner()
image_helper = ImageHelper()

tensor_1 = image_helper.image_to_tensor('./datasets/set5/woman.png')
tensor_2 = image_helper.image_to_tensor('./datasets/set5/bird.png')
```

```python
# Define the Gaussian noise parameters (mean and standard deviation)
mean = 0.0
std = 0.1  # You can adjust the standard deviation as needed

# Create an instance of the AddGaussianNoise transform
add_noise = AddGaussianNoise(mean, std)

# Apply the noise transform to tensor_1
noisy_tensor_1 = add_noise(tensor_1)
noisy_tensor_2 = add_noise(tensor_2)

# # Ensure the noisy_tensor_1 is within a valid range (0 to 1 for typical images)
# noisy_tensor_1 = torch.clamp(noisy_tensor_1, 0.0, 1.0)
```

```python
print('tensor_1 ssim =', model_runner.calculate_ssim(tensor_1, noisy_tensor_1))
print('tensor_1 psnr =', model_runner.calculate_psnr(tensor_1, noisy_tensor_1))
print('tensor_2 ssim =', model_runner.calculate_ssim(tensor_2, noisy_tensor_2))
print('tensor_2 psnr =', model_runner.calculate_psnr(tensor_2, noisy_tensor_2))
```
